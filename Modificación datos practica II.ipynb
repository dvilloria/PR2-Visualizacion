```python
data2 = data.copy()
data2['Age'] = data2['Age'].round()

# Crear grupos de edad
bins = [13, 24, 34, 44, 54, 100]
labels = ['14-24', '25-34', '35-44', '45-54', '55-64']
data2['Age_group'] = pd.cut(data2['Age'], bins=bins, labels=labels, right=False)
```

```python

data['Age'] = data['Age'].round()

# Crear grupos de edad
bins = [13, 24, 34, 44, 54, 100]
labels = ['14-24', '25-34', '35-44', '45-54', '55-64']
data['Age_group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)
```

```python
edad = data.groupby(['Age_group', 'NObeyesdad']).size().reset_index(name='count')

# Pivotar los datos para obtener el formato deseado
pivot_edad = edad.pivot_table(index='Age_group', columns='NObeyesdad', values='count', aggfunc='sum')
```

```python
count_df = data.groupby(['Age_group', 'NObeyesdad']).size().reset_index(name='count')
# Calcular el total de registros por grupo de edad
age_group_totals = count_df.groupby('Age_group')['count'].sum().reset_index(name='total')

# Unir el total de cada grupo de edad con el DataFrame de conteo
count_df = count_df.merge(age_group_totals, on='Age_group')

# Calcular el porcentaje dentro de cada grupo de edad
count_df['percentage'] = (count_df['count'] / count_df['total']) * 100

# Pivotar los datos para obtener el formato deseado
pivot_df = count_df.pivot_table(index='Age_group', columns='NObeyesdad', values='percentage', aggfunc='sum')

# Resetear el índice para una visualización más clara
pivot_df = pivot_df.reset_index()
```

```python
data2['FAF'] = data2['FAF'].round()
```

```python
deporte = data2.groupby(['Age_group', 'FAF']).size().reset_index(name='count')
# Calcular el total de registros por grupo de edad
age_group_totals = deporte.groupby('Age_group')['count'].sum().reset_index(name='total')

# Unir el total de cada grupo de edad con el DataFrame de conteo
deporte = deporte.merge(age_group_totals, on='Age_group')

# Calcular el porcentaje dentro de cada grupo de edad
deporte['percentage'] = (deporte['count'] / deporte['total']) * 100

# Pivotar los datos para obtener el formato deseado
pivot_depor = deporte.pivot_table(index='Age_group', columns='FAF', values='percentage', aggfunc='sum')

# Resetear el índice para una visualización más clara
pivot_depor = pivot_depor.reset_index()
```

```python
obesos = data[data['NObeyesdad'].str.contains('Obesity')]
```

```python
noobesos = data[~data['NObeyesdad'].str.contains('Obesity')]
```

```python
obesidad_transport_counts = obesos['MTRANS'].value_counts() / len(obesos) * 100
no_obesidad_transport_counts = noobesos['MTRANS'].value_counts() / len(noobesos) * 100
```

```python
obesidad_transport_counts
```

```python
no_obesidad_transport_counts
```

```python
df_obesidad_porcentaje = pd.DataFrame(obesidad_transport_counts).rename(columns={'MTRANS': 'Obesos'})
df_no_obesidad_porcentaje = pd.DataFrame(no_obesidad_transport_counts).rename(columns={'MTRANS': 'No Obesos'})

# Paso 4: Unir los DataFrames
df_porcentaje = pd.concat([df_obesidad_porcentaje, df_no_obesidad_porcentaje], axis=1)

# Visualizar el DataFrame resultante
print(df_porcentaje)
```

```python
df_porcentaje.to_csv('obesity_transport.csv')
```

```python
# Paso 2: Calcular los recuentos y porcentajes para obesos
counts_obesos = obesos['family_history_with_overweight'].value_counts()
percentages_obesos = obesos['family_history_with_overweight'].value_counts(normalize=True) * 100

# Paso 3: Calcular los recuentos y porcentajes para no obesos
counts_no_obesos = noobesos['family_history_with_overweight'].value_counts()
percentages_no_obesos = noobesos['family_history_with_overweight'].value_counts(normalize=True) * 100

# Paso 4: Crear DataFrames para visualizar los resultados
results_obesos_df = pd.DataFrame({'Recuento Obesos': counts_obesos, 'Porcentaje Obesos': percentages_obesos})
results_no_obesos_df = pd.DataFrame({'Recuento No Obesos': counts_no_obesos, 'Porcentaje No Obesos': percentages_no_obesos})

# Paso 5: Unir los resultados en un solo DataFrame
results_df = pd.concat([results_obesos_df, results_no_obesos_df], axis=1)

# Mostrar el DataFrame
print(results_df)
```

```python
results_df.to_csv('obesity_historic.csv')
```
